```python
import math
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# global constant for net inputs (can be easily changed)
NET_INPUTS = np.linspace(-10, 10, 100)
```

```python
# function to graph the plots
def show_graph(title, activations, label):
    plt.title(title)
    plt.xticks([-10, -8, -6, -4, -2, 0, 2, 4, 6, 8, 10])
    plt.xlabel("net input")
    plt.ylabel("activation")
    plt.plot(NET_INPUTS, activations, label=label)
```

<b>Question 1, part a</b>

```python
# logistic activation function that uses a loop
def logistic_func_loop(n):
    x = np.zeros(100)
    for i in range(len(n)):
        val = 1/(1+math.exp(-n[i]))
        x[i] = val
    return x
```

```python
# logistic activation function that doesn't use a loop
def logistic_func_vectorized(n):
    return 1/(1+np.exp(-n))
```

<b>Question 1, part b</b>

```python
# relu activation function that uses a loop
def relu_func_loop(n):
    x = np.zeros(100)
    for i in range(len(n)):
        if n[i] >= 0:
            x[i] = n[i]
    return x
```

```python
# relu activation function that doesn't use a loop
def relu_func_vectorized(n):
    x = tf.keras.activations.relu(n)
    return x
```

<b>Question 1, part c</b>

```python
show_graph("Logistic Function, For Loop", logistic_func_loop(NET_INPUTS), "")
plt.show()
```

```python
show_graph("Logistic Function, Vectorized", logistic_func_vectorized(NET_INPUTS), "")
plt.show()
```

```python
show_graph("Relu Function, For Loop", relu_func_loop(NET_INPUTS), "")
plt.show()
```

```python
show_graph("Relu Function, Vectorized", relu_func_vectorized(NET_INPUTS), "")
plt.show()
```

<b>Question 1, part d</b>

```python
a = np.linspace(-3, 3, 100)
# fill in input arrays for different weights, holding the scalar at constant 0
n1 = a*20+0 # w=20, b=0
n2 = a*10+0 # w=10, b=0
n3 = a*1+0 # w=1, b=0
```

```python
show_graph("Logistic Function For Different Values of W", logistic_func_vectorized(n1), label="W=20")
show_graph("Logistic Function For Different Values of W", logistic_func_vectorized(n2), label="W=10")
show_graph("Logistic Function For Different Values of W", logistic_func_vectorized(n3), label="W=1")
plt.legend()
plt.show()
```

```python
show_graph("Relu Function For Different Values of W", relu_func_vectorized(n1), label="W=20")
show_graph("Relu Function For Different Values of W", relu_func_vectorized(n2), label="W=10")
show_graph("Relu Function For Different Values of W", relu_func_vectorized(n3), label="W=1")
plt.legend()
plt.show()
```

```python
# fill in input arrays for different slopes, holding the weights at constant 10
n1 = a*10-10 # w=10, b=-10
n2 = a*10+10 # w=10, b=10
n3 = a*10+20 # w=10, b=20
```

```python
show_graph("Logistic Function For Different Values of b", logistic_func_vectorized(n1), label="b=-10")
show_graph("Logistic Function For Different Values of b", logistic_func_vectorized(n2), label="b=10")
show_graph("Logistic Function For Different Values of b", logistic_func_vectorized(n3), label="b=20")
plt.legend()
plt.show()
```
